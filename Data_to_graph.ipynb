```python
# matplotlib display!! for only one L 
# this is how we put x, y into the graph. we can do so also for plotly. important just to generate data.

def plot_matplotlib(x, y, L, title):
    plt.plot(x, y, label = L, linewidth=2)
    plt.legend(title="Lattice dimension", loc=2, fontsize='small', fancybox=True)
    plt.ylabel('probability of decoder failure')
    plt.xlabel('probability of error per qubit')
    plt.title(title, size=20);
```

```python
def plot_plotly(x, y, d_range):
    fig = go.Figure()
    for d in d_range:
        fig.add_trace(go.Scatter(x=x, y=y[d], mode='lines+markers', name=d))
    fig.update_layout(autosize=False,
    width=500,
    height=500,
    xaxis_title='probability of error per qubit',
    yaxis_title='probability of decoder failure')
    return fig
```
